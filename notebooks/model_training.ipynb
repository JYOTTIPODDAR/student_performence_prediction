5.FEATURE ENGINEERING

COPY DATA


```python
df_clean = df_clean.copy()
```

CREATE ACADEMIC INTENSITY FEATURE

```python
df_clean["study_effort"] = (
    df_clean["Hours_Studied"] * 0.6 +
    df_clean["Attendance"] * 0.4
)
```

CREATE STUDENT CONSISTENCY FEATURE

```python
df_clean["academic_consistency"] = (
    df_clean["Previous_Scores"] +
    df_clean["study_effort"]
) / 2
```

CREATE LIFESTYLE BALANCE FEATURE

```python
df_clean["lifestyle_balance"] = (
    df_clean["Sleep_Hours"] +
    df_clean["Physical_Activity"]
) / 2
```

Attendance level

```python
df_clean["attendance_level"] = pd.cut(
    df_clean["Attendance"],
    bins=[0, 60, 80, 100],
    labels=["low", "medium", "high"]
)
```

ENCODE NEW CATEGORICAL FEATURE

```python
df_clean = pd.get_dummies(
    df_clean,
    columns=["attendance_level"],
    drop_first=True
)
```

DROP REDUNDANT / LESS USEFUL FEATURES

```python
df_clean.drop(
    ["Hours_Studied", "Attendance"],
    axis=1,
    inplace=True
)
```

DATA SPLITTING

```python
# reload clean dataframe (fresh start)
df_temp = df.copy()
```

```python
threshold = df_temp["Exam_Score"].median()

df_temp["pass_fail"] = (df_temp["Exam_Score"] >= threshold).astype(int)
```

```python
df_temp["pass_fail"].value_counts()
```

```python
df_temp.drop("Exam_Score", axis=1, inplace=True)
```

```python
X = df_temp.drop("pass_fail", axis=1)
y = df_temp["pass_fail"]
```

```python
import pandas as pd
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_rf))
```

```python
sample = X_test[0].reshape(1, -1)

pred = rf.predict(sample)

if pred[0] == 1:
    print("Student is likely to PASS")
else:
    print("Student is likely to FAIL")
```
